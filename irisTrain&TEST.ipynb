```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# Import dataset
df = pd.read_csv("/content/Iris.csv")
#df=f.drop('Id',axis=1,inplace=True)
#lst = df.values.tolist()
#l= [tuple(x) for x in lst]
#print(l)
df['Species'].unique()


```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['Species']= label_encoder.fit_transform(df['Species'])
df['Species'].unique()
#lst = df.values.tolist()
#new_lst = [tuple(x) for x in lst]
#print(new_lst)
#print(df['Species'])
```

```python
tr_dt = df.sample(frac=0.8, random_state=25)
te_dt = df.drop(tr_dt.index)

print(f"No. of training examples: {tr_dt.shape[0]}")
print(f"No. of testing examples: {te_dt.shape[0]}")
```

```python
w1=[0.2,0.3,0.4,0.1]
w2=[0.1,0.3,0.2,0.6]
w=[w1,w2]
w3=[0.3,0.2]
b=0.2
qe=[0]*150
p=0
#p=0
he=[0,0]
def linwe(w,x,b):
  p=np.dot(w,x)
  return (p+b)

def act(a):
  z = 1/(1 + np.exp(-a))
  return(z)
def oerr(o,u):
   oe= o*(1-o)*(u-o)
   return(oe)
def herr(h,u,w):
  for j in range(len(h)):
     he[j]=h[j]*(1-h[j])*u*w[j]
  return(he)
def wtupdt(w,n,e,x):
 # for j in range(len(e)):
      for i in range(len(w)):
         w[i]=w[i]+(n*e*x[i])
      return(w)

for r in range(10):
  print("EPOCH NUMBER =",r)
  for c in range(len(tr_dt)):
    m=[tr_dt.iloc[c, 1], tr_dt.iloc[c, 2],tr_dt.iloc[c, 3],tr_dt.iloc[c, 4]]
    o1= linwe(w1,m,b)
    o2= linwe(w2,m,b)
    h1= act(o1)
    h2= act(o2)
    k=[h1,h2]
    print("TUPLE NUMBER---",c)
    print("o1=",o1)
    print("o2=",o2)
    print("h1=",h1)
    print("h2=",h2)
    print("LAST LAYER:")
    q=linwe(w3,k,b)
    y=act(q)
    print("FINAL PREDICTED OUTPUT =",math.floor(y))
    qe[c]=oerr(q,tr_dt.iloc[c,5])

  for c in range(len(qe)):
    p=p+qe[c]
  E=p/len(qe)
  hier=herr(k,E,w3)


  print("Corresponding Error Term =",E)
  print("Hidden-Output Weight Update=",wtupdt(w3,0.5,E,k))
  print("Hidden layer Error term =",hier)
  print("Input-Hidden Weight Updates:-")
  #for s in range(len(hier)):
  a1=wtupdt(w1,0.5,hier[0],m)
  a2=wtupdt(w2,0.5,hier[1],m)
  print(" ",a1)
  print(" ",a2)


















```

```python
from sklearn.metrics import accuracy_score


n=[0]*len(te_dt)
g=[0]*len(te_dt)
for c in range(len(te_dt)):
    m=[te_dt.iloc[c, 1], te_dt.iloc[c, 2],te_dt.iloc[c, 3],te_dt.iloc[c, 4]]
    o1= linwe(w1,m,b)
    o2= linwe(w2,m,b)
    h1= act(o1)
    h2= act(o2)
    k=[h1,h2]
    print("TUPLE NUMBER---",c)
    print("o1=",o1)
    print("o2=",o2)
    print("h1=",h1)
    print("h2=",h2)
    print("LAST LAYER:")
    q=linwe(w3,k,b)
    n[c]=math.floor(act(q))
    g[c]=te_dt.iloc[c,5]
    print("FINAL PREDICTED OUTPUT =",n[c])
    print("ACTUAL TARGET VALUE =",te_dt.iloc[c,5])


r=accuracy_score(n,g, normalize=False)
print("ACCURACY_SCORE =",r)


```
